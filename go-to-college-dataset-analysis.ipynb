```python
OHE = pd.get_dummies(student.iloc[:, [0, 1, 2, 3, 4]])
drop_student = student.drop(['type_school', 'school_accreditation', 'gender', 'interest', 'residence'], axis = 1)
```

```python
OHE
```

```python
drop_student
```

```python
student2 = pd.concat([OHE, drop_student], axis = 1)
```

```python
student2
```

```python
x_train,x_test,y_train,y_test=train_test_split(student2, in_college, test_size=0.3, random_state=46)
print(x_train.shape,x_test.shape)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
print(knn.score(x_train,y_train))
print(knn.score(x_test,y_test))
```

```python
train_accuracy = []
test_accuracy = []

max_train_accuracy = 0
train_index = 0
max_test_accuracy = 0
test_index = 0

for i in range(28):
    knn=KNeighborsClassifier(n_neighbors=i+3)
    knn.fit(x_train,y_train)
    train_score = knn.score(x_train,y_train)
    train_accuracy.append(train_score)
    test_score = knn.score(x_test,y_test)
    test_accuracy.append(test_score)
    
    if train_score > max_train_accuracy:
        max_train_accuracy = train_score
        train_index = i+3
    if test_score > max_test_accuracy:
        max_test_accuracy = test_score
        test_index = i+3   
print("train_score의 최댓값: ", max_train_accuracy, " ", train_index, "번째")
print("test_score의 최댓값: ", max_test_accuracy, " ", test_index, "번째")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

features = scaler.fit_transform(student2)
```

```python
x_train,x_test,y_train,y_test=train_test_split(features, in_college, test_size=0.3, random_state=46)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```
